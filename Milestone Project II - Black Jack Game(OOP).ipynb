## Playing Cards
A standard deck of playing cards has four suits (Hearts, Diamonds, Spades and Clubs) and thirteen ranks (2 through 10, then the face cards Jack, Queen, King and Ace) for a total of 52 cards per deck. Jacks, Queens and Kings all have a rank of 10. Aces have a rank of either 11 or 1 as needed to reach 21 without busting.

```python
import random
from IPython.display import clear_output
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing=True
```

```python
class Card:
    
    def __init__(self,club,rank):
        self.club=club
        self.rank=rank
        
    def __str__(self):
        #This Method uses constructor and print the card in word format
        return "{} of {}".format(self.rank,self.club)
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # empty list to store all cards in deck
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        st="The deck has:"
        for i in self.deck:
            st+="\n\t"+str(i)
        return st
    def shuffle(self):
        random.shuffle(self.deck)
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # empty list to store cards
        self.value = 0  
        self.aces = 0    
    
    def adjust_for_ace(self):
        if self.aces>0:
            self.value+=(self.aces-1)
            if self.value+11>21:
                self.value+=1
            else:
                self.value+=11
    
    def add_card(self,card):
        self.cards.append(card)
        self.value=0
        self.aces=0
        for i in self.cards:
            if i.rank!="Ace":
                self.value+=values[i.rank]
            else:
                self.aces+=1
    def __str__(self):
        for i in self.cards:
            print("\t",i)
        for g in range(1):
            return ""
        
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100  
        self.bet = 0
        
    def win_bet(self):
        self.total+=self.bet
    
    def lose_bet(self):
        self.total-=self.bet
```

```python
def hit(dec,hand):
    hand.add_card(dec.deck[0])
    hand.adjust_for_ace()
    dec.deck=dec.deck[1:]
```
